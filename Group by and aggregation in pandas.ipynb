```python
df.groupby('Base Flavor').mean()
```

```python
# to get the count of base flavour 
```

```python
df.groupby('Base Flavor').count()
```

```python
# to get the minimum value of base flavour 
```

```python
df.groupby('Base Flavor').min()
```

```python
# to get the maximum value of base flavour 
```

```python
df.groupby('Base Flavor').max()
```

```python
# to calculate the sum value of base flavour 
```

```python
df.groupby('Base Flavor').sum()
```

```python
# or i can use the aggregation function to get mean, max, count and sum of base flavour flavour rating
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'max', 'count', 'sum']})
```

```python
# i can also use the aggregation function to get mean, max, count and sum of base flavour for flavour rating and texture rating
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'max', 'count', 'sum'], 'Texture Rating': ['mean', 'max', 'count', 'sum']})
```
